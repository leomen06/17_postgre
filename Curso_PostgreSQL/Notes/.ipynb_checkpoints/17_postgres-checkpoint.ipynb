### <a name="clase-2-Que-es-Postgresql"></a>Clase 2 Que es Postgresql

### [Index](#index)

Postgre es un motor de base de datos open source y usa el Objeto-Relacional, existen tres conceptos importantes acerca de las bases de datos

- lenguaje
- motor: Permite estructurar toda la información dentro de un servidor
- servidor

Esta base de datos cumple con el estándar ACID (reglas para bases de datos para tener buenas practicas)

**• A: Atomicity – Atomicidad** -> Separar las funciones desarrolladas en la BD como pequeñas tareas y ejecutarlas como un todo. Si alguna tarea falla se hace un rollback(Se deshacen los cambios)

**• C: Consistency – Consistencia** -> Todo lo que se desarrolló en base al objeto relacional. Los datos tienen congruencia (gracias a PK y FK)

**• I: Isolation – Aislamiento** -> Varias tareas ejecutándose al mismo tiempo dentro de la BD (o varias instancias)

**• D: Durability – Durabilidad** -> Puedes tener seguridad que la información no se perderá por un fallo catastrófico. PostgreSQL guarda la información en una Bitácora y luego guarda el cambio efectivo en la base de datos.

Por Que PostgreSQL

- Tipos de Datos
- Integridad de Datos
- Concurrencia, Rendimiento
- Fiabilidad, recuperación ante desastres
- Seguridad
- Extensibilidad
- Internacionalizacion
- Búsqueda de Texto
- Funciones como PostGIS y PL

## Stack Builder en PostgreSQL

**Stack Builder** es una herramienta gráfica que facilita la instalación y gestión de módulos adicionales para PostgreSQL. Se incluye con el instalador oficial de PostgreSQL de EDB (EnterpriseDB) y se puede utilizar para:

* **Instalar módulos**: Stack Builder te permite descargar e instalar una amplia gama de módulos complementarios para PostgreSQL, como pgAdmin, PostGIS, pgBouncer, y muchos más.
* **Gestionar dependencias**: Stack Builder se encarga automáticamente de resolver las dependencias entre módulos, lo que te evita tener que instalar manualmente los paquetes necesarios.
* **Actualizar módulos**: Stack Builder te permite actualizar los módulos instalados a la última versión disponible.
* **Eliminar módulos**: Puedes eliminar fácilmente los módulos que ya no necesitas.

**Ventajas de usar Stack Builder:**

* **Fácil de usar**: Stack Builder tiene una interfaz gráfica intuitiva que facilita su uso, incluso para usuarios sin experiencia técnica.
* **Ahorra tiempo**: Stack Builder automatiza el proceso de instalación y gestión de módulos, lo que te permite ahorrar tiempo y esfuerzo.
* **Reduce errores**: Stack Builder se encarga de resolver las dependencias entre módulos, lo que reduce el riesgo de errores durante la instalación.

**Cómo usar Stack Builder:**

1. **Abre Stack Builder**: Puedes encontrar Stack Builder en el menú de inicio de Windows o en la carpeta de aplicaciones de macOS.
2. **Selecciona una instalación de PostgreSQL**: Stack Builder te mostrará una lista de las instalaciones de PostgreSQL que tienes en tu ordenador. Selecciona la instalación con la que deseas trabajar.
3. **Selecciona el módulo que deseas instalar**: En la pestaña "Módulos", puedes buscar y seleccionar el módulo que deseas instalar.
4. **Haz clic en "Instalar"**: Stack Builder descargará e instalará el módulo automáticamente.
5. **(Opcional) Actualiza o elimina módulos**: Puedes actualizar o eliminar módulos instalados desde la pestaña "Módulos".

**Recursos adicionales:**

* **Documentación de Stack Builder:** [https://www.enterprisedb.com/docs/postgresql_journey/02_installing/01_EDB_installer/03_using_stackbuilder/](https://www.enterprisedb.com/docs/postgresql_journey/02_installing/01_EDB_installer/03_using_stackbuilder/)

### <a name="clase-6-archivos-de-configuracion"><a/> Clase 6 Archivos de Configuracion 
    
[Index](#index)

Antes de crear tablas o hacer diseños es importante conocer 3 archivos de configuracion que muchas veces son el origen de los problemas cuando estamos agregando nuevos servicios o configurando la base de datos

Los archivos de configuración son tres principales:

- **postgreql.conf**: en este archivo podemos cambiar puerto, numero de conexiones maxima, accesos para ip únicas. (un pro en bases de datos siempre ha leído todo el documento). En este archivo también se encuentra la información de las replicas

- **pg.hba.conf**: contiene los roles, tipos de usuarios, permisos, direcciones, métodos de acceso con MD5, también puedo negar a otros hosts a conectarnos,este archivo es muy importante dado que nos permite no solo dar acceso a usuarios y a las replicas sino protegernos de ataques o personas mal intencionados

- **pg_ident.conf**: El archivo pg_ident.conf es un archivo de configuración opcional en PostgreSQL que se utiliza para definir la relación entre los nombres de usuario del sistema operativo y los nombres de usuario de la base de datos. En otras palabras, permite mapear los usuarios del sistema a usuarios específicos dentro de PostgreSQL.

La ruta de los mismos depende del sistema Operativo, para saber que que ruta están, basta con hacer una Query

**SHOW config_file**;

NOTA: siempre es bueno hacer una copia original de los archivos antes de modificarlos como un respaldo.

### <a name="clase-7-comandos-mas-utilizados-en-postgresql"><a/>Clase 7 Comandos mas utilizados en PostgreSQL
    
[Index](#index)

La consola en PostgreSQL es una herramienta muy potente para crear, administrar y depurar nuestra base de datos. podemos acceder a ella después de instalar PostgreSQL y haber seleccionado la opción de instalar la consola junto a la base de datos.

PostgreSQL está más estrechamente acoplado al entorno UNIX que algunos otros sistemas de bases de datos, utiliza las cuentas de usuario nativas para determinar quién se conecta a ella (de forma predeterminada). El programa que se ejecuta en la consola y que permite ejecutar consultas y comandos se llama psql, psql es la terminal interactiva para trabajar con PostgreSQL, es la interfaz de línea de comando o consola principal, así como PgAdmin es la interfaz gráfica de usuario principal de PostgreSQL.

Después de emitir un comando PostgreSQL, recibirás comentarios del servidor indicándote el resultado de un comando o mostrándote los resultados de una solicitud de información. Por ejemplo, si deseas saber qué versión de PostgreSQL estás usando actualmente, puedes hacer lo siguiente:

```sql
SELECT version()

```

Para crear una base de datos debes ejecutar la consulta de creación de base de datos, es importante entender que existe una costumbre no oficial al momento de escribir consultas; consiste en poner en mayúsculas todas las palabras propias del lenguaje SQL cómo **CREATE**, **SELECT**, **ALTER**, etc y el resto de palabras como los nombres de las tablas, columnas, nombres de usuarios, etc en minúscula. ### <a name="clase-13-particiones"><a/>Clase 13 Particiones  [Index](#index)

En ocasiones llega un punto en el que tienes mucha informacion en una sola tabla, y al momento de consultarla la base de datos tendrá que hacer el recorrido de toda la informacion para encontrar el contenido que estas buscando.

Imagina que nuestra tabla de viajes tiene cien mil registros, y que quieres consultar los viajes realizados en un rango de fechas, esa consulta ocupara un espacio en memoria que deberá ser leída para obtener la respuesta, en este punto entran las particiones.

Las particiones tiene como ventaja lo siguiente:

- Separación física de datos
- Conservan la Estructura Lógica

La separacion física hace referencia a que es posible guardar varias parte de la misma tabla en diferentes espacios de disco o incluso en otros discos, internamente Postgre usando el mismo nombre de la tabla asigna otras pequeñas tablas que tienen un rango definido, en nuestro caso le diríamos que el rango son las fechas o las horas, por ejemplo una partición para junio, otra para julio y al hacer la consulta con el mismo **SELECT** de siempre Postgre buscaría la informacion solo en la partición correspondiente, esto reduce el tiempo de ejecucion de la consulta y a su vez evita que la tabla se bloquee al ser consultada o escrita demasiadas veces en periodos de tiempo cortos.

En esta clase vamos a crear una tabla, asignarle particiones y después hacer un **INSERT** común y corriente y hacer una consulta común y corriente con pgAdmin.

### Comentarios_01:
    
`************************************************************************************************************************`

Supongamos que tienes dos tablas: `Ventas` y `Clientes`. La tabla `Ventas` está particionada por año, y ambas tablas tienen una columna llamada `ClienteID` que se puede utilizar para realizar un JOIN. Aquí está el ejemplo:
    
```sql
-- Crear la tabla de Clientes
CREATE TABLE Clientes (
    ClienteID SERIAL PRIMARY KEY,
    Nombre VARCHAR(50)
);

-- Crear la tabla de Ventas particionada por año
CREATE TABLE Ventas (
    VentaID SERIAL PRIMARY KEY,
    ClienteID INT,
    Monto DECIMAL(10, 2),
    Fecha DATE,
    CONSTRAINT fk_cliente FOREIGN KEY (ClienteID) REFERENCES Clientes(ClienteID)
) PARTITION BY RANGE (EXTRACT(YEAR FROM Fecha))) (
    START (2000) END (2010) INCLUSIVE,
    START (2011) END (2020) INCLUSIVE,
    START (2021) END (MAXVALUE)
);

-- Insertar datos de ejemplo en ambas tablas
INSERT INTO Clientes (Nombre) VALUES ('Cliente A');
INSERT INTO Ventas (ClienteID, Monto, Fecha) VALUES
(1, 100.00, '2005-05-10'),
(1, 150.50, '2012-08-20');

-- Realizar una operación JOIN entre las tablas
SELECT Ventas.VentaID, Ventas.Monto, Ventas.Fecha, Clientes.Nombre
FROM Ventas
JOIN Clientes ON Ventas.ClienteID = Clientes.ClienteID;
    
```
  
Este ejemplo crea dos tablas (Clientes y Ventas) y realiza un JOIN entre ellas utilizando la columna ClienteID. La tabla Ventas está particionada por año. Aunque es un ejemplo básico, ilustra que la partición no impide la existencia de una clave primaria ni afecta la capacidad de realizar JOIN con otras tablas.
    
No, en el diseño actual, los valores de `VentaID` no se repetirán entre las particiones. La columna `VentaID` es la clave primaria de la tabla `Ventas`, y al ser de tipo `SERIAL` en PostgreSQL, automáticamente se incrementará de forma única para cada fila insertada en la tabla, independientemente de la partición.

En otras palabras, cada partición contendrá un conjunto único de valores de `VentaID`, y no habrá duplicados en diferentes particiones. La columna `VentaID` actuará como una secuencia independiente para cada partición.

Hasta ahora hemos estado utilizando el usuario predeterminado de postgres, lo cual sabemos que es muy peligroso ya que al ser un usuario de tipo root este tiene todos los privilegios para borrar todo, lo ideal es crear un usuario para trabajar con los roles y permisos adecuados.

#### <a name="creando-roles-desde-sql-shell"><a/>Creando roles desde SQL Shell [Index](#index)

Crearemos un rol con la terminal que tenga la capacidad de hacer login

```sql
\H CREATE ROLE

CREATE ROLE nombre [ [ WITH ] opción [ ... ] ]

donde opción puede ser:

 SUPERUSER | NOSUPERUSER
| CREATEDB | NOCREATEDB
| CREATEROLE | NOCREATEROLE
| INHERIT | NOINHERIT
| LOGIN | NOLOGIN
| REPLICATION | NOREPLICATION
| BYPASSRLS | NOBYPASSRLS
| CONNECTION LIMIT limite_conexiones
| [ ENCRYPTED ] PASSWORD 'contraseña' | PASSWORD NULL
| VALID UNTIL 'fecha_hora'
| IN ROLE nombre_de_rol [, ...]
| IN GROUP nombre_de_rol [, ...]
| ROLE nombre_de_rol [, ...]
| ADMIN nombre_de_rol [, ...]
| USER nombre_de_rol [, ...]
| SYSID uid
```

CREATE USER es un alias de CREATE ROLE

```sql
postgres=# CREATE ROLE usuario_consulta;
CREATE ROLE
postgres=# \dg
Lista de roles
  Nombre de rol   | Atributos | Miembro de
------------------+------------------------------------------------------------+------------
 postgres| Superusuario, Crear rol, Crear BD, Replicacion, Ignora RLS | {}
 usuario_consulta | No puede conectarse | {}


postgres=#
```

En este punto hemos creado un nuevo usuario/role, pero este no tiene atributos/permisos para realizar ninguna acción, solucionamos esto igual que una tabla con el comando ALTER.

```sql
postgres=# ALTER ROLE  usuario_consulta WITH LOGIN;
ALTER ROLE


postgres=# \dg
Lista de roles
  Nombre de rol   | Atributos | Miembro de
------------------+------------------------------------------------------------+------------
 postgres| Superusuario, Crear rol, Crear BD, Replicacion, Ignora RLS | {}
 usuario_consulta | | {}

```

Observamos ahora que desaparece la leyenda de que el usuario no puede conectarse, ahora realizamos la misma operación para darle los atributos de un superusuario.

```SQL
postgres=# ALTER ROLE usuario_consulta SUPERUSER;
ALTER ROLE
postgres=# \dg
Lista de roles
  Nombre de rol   | Atributos | Miembro de
------------------+------------------------------------------------------------+------------
 postgres| Superusuario, Crear rol, Crear BD, Replicacion, Ignora RLS | {}
 usuario_consulta | Superusuario  | {}
```

- **Mantenimiento manual:** Es necesario realizar una validación manual posterior para garantizar la integridad referencial.

**Recomendaciones:**

- **Utilizar con precaución:** Se recomienda usar "NOT VALID" solo cuando exista una razón justificada para evitar la validación inicial.
- **Validar manualmente:** Es importante validar manualmente los datos existentes para asegurar la integridad referencial.

Los siguientes son los tipos de joins

lectura recomendada  <https://www.postgresqltutorial.com/postgresql-joins/>

Consultas de la clase

```sql
-- Inner Join o join
select * from pasajero
join viaje on (viaje.id_pasajero = pasajero.id);

-- que pasajeros no han tomado ningun viaje
select * from pasajero
LEFT JOIN viaje on (viaje.id_pasajero = pasajero.id)
WHERE viaje.id IS NULL;
```

![join_](src/join_1.jpg)
    
`************************************************************************************************************************`
    
No es necesario tener una PK y/o FK para poder hacer un join. Un join es una operación que combina filas de dos o más tablas en función de una relación entre ellas. Esta relación puede ser establecida mediante una PK, una FK, o simplemente mediante una columna común.
    
### Eficiencia:
    
En términos de eficiencia, definir una PK y FK es mejor que hacer un join con tablas que no lo tengan definidos, suponiendo que en cada tabla hay una columna que las puede vincular, por las siguientes razones:

- La PK y FK definen la relación entre las tablas de forma explícita. Esto significa que el motor de base de datos puede utilizar esta información para realizar el join de forma más eficiente.

- La PK y FK garantizan que los valores de las columnas relacionadas sean únicos. Esto evita que se produzcan errores en las consultas que utilizan el join.
    
En el caso de un join con tablas que no tienen PK y FK, el motor de base de datos tendrá que utilizar un algoritmo más general para encontrar las filas coincidentes. Este algoritmo puede ser menos eficiente que un join basado en una PK y/o FK, ya que tiene que comparar todos los valores de las columnas relacionadas.
    
### Algoritmos:
    
**En el caso de un join con tablas que tienen PK y FK, el motor de base de datos utiliza uno de los siguientes algoritmos:**

* **Hash join:** Este algoritmo utiliza un hash table para almacenar los valores de la columna primaria de la tabla de origen. Luego, recorre la tabla de destino, comparando los valores de la columna relacionada con los valores de la tabla hash.
* **Merge join:** Este algoritmo ordena las dos tablas por la columna relacionada. Luego, recorre las dos tablas en paralelo, comparando los valores de la columna relacionada.

**En el caso de un join con tablas que no tienen PK y FK, el motor de base de datos utiliza uno de los siguientes algoritmos:**

* **Nested loop join:** Este algoritmo recorre la tabla de origen una vez, comparando los valores de la columna relacionada con los valores de la tabla de destino.
* **Block nested loop join:** Este algoritmo es similar al nested loop join, pero utiliza bloques de datos para mejorar el rendimiento.

**El algoritmo que utiliza el motor de base de datos depende de varios factores, como el tamaño de las tablas, el orden de las tablas y el tipo de datos de las columnas relacionadas.**

**En general, el hash join es el algoritmo más eficiente para joins con tablas grandes y bien indexadas.** El merge join es un buen compromiso entre eficiencia y rendimiento. El nested loop join es el algoritmo menos eficiente, pero es el único que puede utilizarse cuando las tablas no están indexadas.

**A continuación, se muestra una tabla resumen de los algoritmos de join:**

| Algoritmo | Descripción | Ventajas | Desventajas |
|---|---|---|---|
| Hash join | Utiliza un hash table para almacenar los valores de la columna primaria de la tabla de origen. | Eficiencia | Requiere espacio adicional para la hash table |
| Merge join | Ordena las dos tablas por la columna relacionada. | Eficiencia | Requiere ordenar las tablas |
| Nested loop join | Recorre la tabla de origen una vez, comparando los valores de la columna relacionada con los valores de la tabla de destino. | Sencillo | Ineficiente para tablas grandes |

### <a name="clase-19-funciones-especiales-principales"><a/>Clase 19 Funciones Especiales Principales, ON CONFLICT DO, RETURNING, LIKE/ALIKE, IS/IS NOT [Index](#index_01)

Postgres tiene una lista de funciones especiales que están diseñadas exclusivamente para ayudarte a desarrollar tu aplicación mucho mas rápido algunas de ellas son:

- **ON CONFLICT DO:** La cláusula ON CONFLICT DO es una característica específica de PostgreSQL y se utiliza en las sentencias INSERT y UPDATE para manejar conflictos de unicidad en columnas clave o restricciones únicas. Esta cláusula es parte de la funcionalidad de "conflictos en la inserción" o "upsert" (insertar o actualizar) en PostgreSQL.
    
```sql
INSERT INTO nombre_tabla (columna1, columna2, ...)
VALUES (valor1, valor2, ...)
ON CONFLICT (columna_clave)
DO UPDATE SET columna1 = valor1_actualizado, columna2 = valor2_actualizado, ...;
```
    
`ON CONFLICT (columna_clave)`: Esta parte indica qué columnas o restricciones deben considerarse para determinar si hay un conflicto. Puedes especificar la columna clave o una restricción única.

`DO UPDATE SET columna1 = valor1_actualizado, columna2 = valor2_actualizado, ...`: Esto define las acciones a realizar en caso de conflicto. En lugar de lanzar un error, PostgreSQL ejecutará una instrucción UPDATE para modificar las columnas especificadas con los nuevos valores proporcionados.
    
Tipo de acciones:
    
    - NOTHING: No hacer nada y cancelar la operación de INSERT o UPDATE.

    - UPDATE SET ...: Actualizar los valores de las columnas especificadas en las filas que causaron el conflicto.

    - DELETE DO UPDATE SET ...: Eliminar las filas que causaron el conflicto y luego actualizar las filas existentes con los valores especificados.

### <a name="clase-20-funciones-especiales-avanzadas"><a/> Clase 20 Funciones Especiales Avanzadas, COALESCE, NULLIF, GREATEST, LEAST, BLOQUES ANÓNIMOS [Index](#index_01)

Funciones mas comunes en postgres:

### **COALESCE:** 
    
La función `COALESCE` se utiliza en SQL para proporcionar un valor predeterminado cuando se encuentra un valor nulo en una lista de expresiones. La función toma múltiples argumentos y devuelve el primer valor no nulo de la lista. Si todos los valores son nulos, entonces devuelve nulo.

La sintaxis básica de `COALESCE` es la siguiente:

```sql
COALESCE(valor1, valor2, ..., valorN)
```

Aquí hay una explicación de los términos utilizados:

- `valor1, valor2, ..., valorN`: Son las expresiones que se evalúan en orden. `COALESCE` devuelve el primer valor no nulo de esta lista.

Ejemplo:

Supongamos que tienes una tabla de empleados con una columna `nombre` y una columna `apellido`, y algunos registros tienen el apellido como nulo. Puedes usar `COALESCE` para obtener un nombre completo tomando el apellido predeterminado cuando es nulo:

```sql
SELECT nombre, COALESCE(apellido, 'Apellido Desconocido') AS nombre_completo
FROM empleados;
```

En este ejemplo, si el apellido es nulo, se mostrará 'Apellido Desconocido' en lugar de nulo. Si el apellido tiene un valor no nulo, se mostrará ese valor.

En resumen, `COALESCE` es útil para manejar valores nulos y proporcionar un valor predeterminado en su lugar, mejorando la legibilidad y la consistencia de las consultas SQL.

**Ejemplo COALESCE**.

Primero modificar tabla pasajero para tener un dato como nulo.

![funciones_especiales_6](src/funciones_especiales_6.jpg)

```sql
SELECT id, COALESCE(nombre, 'No Aplica') nombre, direccion_residencia, fecha_nacimiento
FROM public.pasajero WHERE id = 1;
```
La función `coalesce()` es una función propia de PostgreSQL. No es parte del estándar SQL, pero es compatible con otras bases de datos como Oracle, SQL Server y MySQL con diferentes nombres.

**¿Qué hace la función coalesce()?**

La función `coalesce()` toma una lista de expresiones como argumentos y devuelve el valor de la primera expresión que no sea `NULL`. Si todas las expresiones son `NULL`, entonces `coalesce()` devuelve `NULL`.

**Ejemplo:**

```sql
SELECT coalesce(nombre, apellido, 'Desconocido');
```

En este ejemplo, la función `coalesce()` devolverá el valor de la columna `nombre` si no es `NULL`. Si `nombre` es `NULL`, la función devolverá el valor de la columna `apellido`. Si `apellido` también es `NULL`, la función devolverá la cadena `"Desconocido"`.

**Sintaxis alternativa:**

La función `coalesce()` también se puede escribir usando la función `CASE WHEN`:

```sql
SELECT CASE WHEN nombre IS NOT NULL THEN nombre
       WHEN apellido IS NOT NULL THEN apellido
       ELSE 'Desconocido'
       END;
```

**Ventajas de usar coalesce():**

* La función `coalesce()` es más concisa y fácil de leer que la sintaxis `CASE WHEN`.
* La función `coalesce()` puede ser más eficiente que la sintaxis `CASE WHEN`.

**Conclusión:**

La función `coalesce()` es una herramienta útil para manejar valores `NULL` en PostgreSQL. Es una función propia de PostgreSQL, pero también es compatible con otras bases de datos con diferentes nombres.

    
### **NULLIF:** 
    
 **En SQL, la función `NULLIF()` se utiliza para devolver un valor nulo si dos expresiones son iguales, o el valor de la primera expresión si no son iguales.** Te permite comparar dos valores y reemplazarlos con un NULL si coinciden, lo que resulta útil para manejar valores nulos y realizar controles condicionales.

**Sintaxis:**

```sql
NULLIF(expression1, expression2)
```

**Argumentos:**

* **expression1:** La primera expresión a evaluar.
* **expression2:** La segunda expresión a evaluar.

**Devuelve:**

* **NULL** si ambas expresiones son iguales.
* **El valor de `expression1`** si las expresiones no son iguales.

**Ejemplos:**

1. **Reemplazar valores nulos:**

   ```sql
   SELECT NULLIF(income, 0) AS adjusted_income
   FROM employees;
   ```
   En este caso, si `income` es 0, la función devolverá NULL, lo que podría ser útil para evitar errores o considerar valores nulos en otros cálculos.

2. **Evitar divisiones por cero:**

   ```sql
   SELECT sales / NULLIF(cost, 0) AS profit_ratio
   FROM products;
   ```
   Si `cost` es 0, la función devuelve NULL, evitando la división por cero y previniendo un error en la consulta.

3. **Comparar valores y devolver un resultado específico:**

   ```sql
   SELECT NULLIF(customer_id, 1001) AS is_special_customer
   FROM orders;
   ```
   Si `customer_id` es igual a 1001, la función devolverá NULL, lo que podría indicar que es un cliente especial y requiere un tratamiento diferente.

**Puntos clave:**

- **Tipos de datos:** Los argumentos `expression1` y `expression2` deben ser del mismo tipo de datos o comparables entre sí.
- **NULL desconocido:** La función trata a NULL como un valor desconocido, no como un valor igual a sí mismo. Por lo tanto, `NULLIF(NULL, NULL)` devuelve NULL.
- **Usos:** La función `NULLIF()` es útil para:
    - Manejar valores nulos de forma controlada.
    - Prevenir errores en cálculos o divisiones por cero.
    - Realizar controles condicionales basados en la comparación de valores.

**En resumen:**

`NULLIF()` es una función versátil para el tratamiento de valores nulos en SQL. Es importante comprender su comportamiento y casos de uso para aprovechar su potencial en consultas y manejo de datos.

    
**Ejemplo NULLIF**.

    
### **GREATEST:** 
    
La función `GREATEST` en SQL sirve para **devolver el mayor valor numérico de una lista de argumentos**. Es útil para comparar múltiples valores y realizar operaciones basadas en el mayor de ellos.

**Sintaxis:**

```sql
GREATEST(expression1, expression2, ...)
```

**Argumentos:**

* `expression1, expression2, ...`: Una lista de expresiones numéricas de cualquier tipo. Pueden ser constantes, columnas de tablas, expresiones matemáticas o incluso otras funciones.

**Devuelve:**

* El valor numérico más grande de la lista de argumentos.

**Ejemplos:**

* Devolver el mayor valor de dos columnas:

```sql
SELECT GREATEST(price, sale_price) AS highest_price
FROM products;
```

* Encontrar el número máximo de una lista de valores constantes:

```sql
SELECT GREATEST(10, 20, 5, 45) AS max_value;
```

* Combinar una expresión con una constante para obtener el mayor de ambos:

```sql
SELECT GREATEST(COALESCE(average_rating, 0), 4) AS overall_rating
FROM reviews;
```

**Observaciones:**

* La función `GREATEST` solo acepta argumentos numéricos. Si un argumento no es numérico, se lo considera como cero.
* Si todos los argumentos son NULL, la función devuelve NULL.
* El orden de los argumentos no afecta el resultado, ya que siempre se busca el mayor valor.

**En resumen:**

`GREATEST` es una función simple y práctica para determinar el mayor valor numérico en una lista de argumentos dentro de tus consultas SQL. Facilita las tareas de comparación y selección de valores máximos, y permite tomar decisiones o realizar operaciones basadas en ellos.


**Ejemplo GREATEST**.

![funciones_especiales_8](src/funciones_especiales_8.jpg)

```sql
SELECT GREATEST (0, 1,2,3,4,5,6);;
```


### **LEAST:** 
    
La función `LEAST` en SQL sirve para **devolver el valor numérico más pequeño de una lista de argumentos**. Es similar a `GREATEST`, pero en lugar de buscar el mayor, encuentra el menor. Esto la vuelve útil para comparar múltiples valores y realizar operaciones basadas en el mínimo.

**Sintaxis:**

```sql
LEAST(expression1, expression2, ...)
```

**Argumentos:**

* `expression1, expression2, ...`: Una lista de expresiones numéricas de cualquier tipo. Pueden ser constantes, columnas de tablas, expresiones matemáticas o incluso otras funciones.

**Devuelve:**

* El valor numérico más pequeño de la lista de argumentos.

**Ejemplos:**

* Encontrar el precio mínimo de un producto:

```sql
SELECT LEAST(price, sale_price) AS lowest_price
FROM products;
```

* Calcular el número de días restantes para una fecha límite:

```sql
SELECT LEAST(DATEDIFF(day, GETDATE(), deadline), 30) AS days_left
FROM tasks;
```

* Combinar una expresión con una constante para obtener el menor de ambos:

```sql
SELECT LEAST(COALESCE(average_rating, 5), 2) AS user_score
FROM reviews;
```

**Observaciones:**

* La función `LEAST` solo acepta argumentos numéricos. Si un argumento no es numérico, se lo considera como infinito positivo.
* Si todos los argumentos son NULL, la función devuelve NULL.
* El orden de los argumentos no afecta el resultado, ya que siempre se busca el menor valor.

**En resumen:**

`LEAST` es una función útil para trabajar con números en tus consultas SQL y encontrar el valor mínimo dentro de una lista de argumentos. Te permite tomar decisiones o realizar operaciones basadas en el menor valor presente.

    
**Ejemplo LEAST**.

### <a name="clase-21-vistas"><a/>Clase 21 Vistas [Index](#index_01)

En PostgreSQL, las vistas y las vistas materializadas son dos conceptos distintos, y cada uno tiene su propósito y características específicas.

1. **Vista Volátil:**
   - Una vista volátil es una consulta predefinida que se almacena en el sistema y se ejecuta cada vez que se referencia.
   - La vista volátil no almacena los resultados de la consulta, sino que genera los resultados en tiempo de ejecución.
   - Cada vez que se consulta la vista, la base de datos ejecuta la consulta subyacente y devuelve los resultados actuales.
   - Las vistas volátiles son útiles para simplificar consultas complejas o proporcionar una capa de abstracción sobre las tablas subyacentes.

   Ejemplo de vista volátil:
   ```sql
   CREATE VIEW vista_ejemplo AS
   SELECT columna1, columna2 FROM tabla WHERE condicion;
   ```

2. **Vista Materializada:**
   - Una vista materializada, por otro lado, almacena físicamente los resultados de la consulta en una tabla especial.
   - La vista materializada se actualiza periódicamente según la configuración (manual o automática) y proporciona resultados más rápidos en consultas repetidas.
   - Pueden ser útiles en situaciones donde la consulta subyacente es costosa en términos de tiempo de ejecución, y se prefiere tener resultados precalculados y actualizados según sea necesario.
   - La desventaja es que la vista materializada ocupa espacio de almacenamiento adicional y puede tener un desfase en la frescura de los datos según la frecuencia de actualización.

   Ejemplo de vista materializada:
   ```sql
   CREATE MATERIALIZED VIEW vista_materializada_ejemplo AS
   SELECT columna1, columna2 FROM tabla WHERE condicion;
   ```

En resumen, las vistas volátiles son consultas ejecutadas en tiempo real, mientras que las vistas materializadas son consultas precalculadas y almacenadas físicamente para un acceso más rápido, con la compensación de la frescura de los datos y el uso de almacenamiento adicional. La elección entre ambas depende de los requisitos específicos y las consideraciones de rendimiento de una aplicación.

Un trigger contiene informacion de lo que se esta cambiando debemos confirmar al motor de la base de datos si el cambio se esta llevando a cabo o no.

Los triggers tiene dos variables globales muy importantes **OLD** lo que estaba antes del cambio y **NEW** lo que esta después del cambio, si queremos que el trigger proceda con el nuevo valor usamos NEW, si queremos quedarnos como estábamos usamos OLD, al igual que la variable rec también podemos acceder a las propiedades con dot notation asi tendremos OLD.id o NEW.nombre donde se almacenan los cambios en el estado, estas tienen mucho sentido cuando hacemos un UPDATE, pero para insert lo adecuado siempre sera NEW.

Seguiremos los siguientes pasos

1 Borramos el Trigger

2.- Borramos la funcion en CASCADE para eliminar el disparador también.

```sql
DROP  FUNCTION public.importante_pl2() CASCADE ;
```

```sql

CREATE FUNCTION public.importante_pl2()
    RETURNS trigger
    LANGUAGE 'plpgsql'
    COST 100
    VOLATILE NOT LEAKPROOF
AS $BODY$
DECLARE
 rec record;
 contador integer := 0;
BEGIN
  FOR rec IN SELECT * FROM pasajero LOOP
    contador := contador + 1;
  END LOOP;
  INSERT INTO cont_pasajero (total,tiempo)
  VALUES (contador, now());
  RETURN NEW;
END
$BODY$;
```

**Solución**.

Funcion pl_on_delete_pasajero

```sql
CREATE FUNCTION public.pl_on_delete_pasajero()
    RETURNS trigger
    LANGUAGE 'plpgsql'

AS $BODY$
DECLARE
 rec record;
 contador integer := 0;
BEGIN
  FOR rec IN SELECT * FROM pasajero LOOP
    contador := contador + 1;
  END LOOP;
  INSERT INTO cont_pasajero (total,tiempo)
  VALUES (contador, now());
  RETURN NEW;
END
$BODY$;
```

### <a name="clase-27-mantenimiento"><a/>Clase 27 Mantenimiento  [Index](#index_01)

Este es un tema importante y que postgres realiza por debajo en segundo plano, en este proceso de vacuum (vaciado) quita todas las filas, columnas e items del disco duro que no están funcionando para optimizar los servicios.

Postgres tiene dos niveles de limpieza.

- Liviano: se ejecuta todo el tiempo para lograr las optimizaciones descritas anteriormente.

- Completo (full): que es capaz de bloquear las tablas para hacer la limpieza y luego las desbloquea, es importante cuando tenemos una tabla muy grande y tenemos problemas de indexacion.

El mantenimiento en PGAdmin lo llevamos a cabo dando click derecho sobre la base de datos o sobre una de las tablas, pero es muy diferente ya que solo afectamos una tabla, mientras a nivel de base de datos podemos llegar a bloquear todas las tablas hasta que la tarea de mantenimiento termine, este es su menu con sus caracteristicas, la mas importante es vacuum.

![maintenance_1](src/maintenance_1.jpg)

- Full: quiere decir que la tabla que vas a limpiar (en este caso estacion) quedara limpia en su totalidad, es decir se realizara la revision de todo el espacio en memoria para que todas las filas que ya no son aplicables se eliminen y todos los indices también.

- Freeze: incluye que durante ese proceso se va a congelar, pero hay que tener en cuenta que un cuando se hace un vacuum full si no es posible hacerlo en lo que llega una consulta nueva esa tabla queda congelada y tendra un lot quiere decir que ninguna tabla podrá acceder a ella hasta que termine el proceso de limpieza.

- Analyze: ejecuta una revision pero no hace cambios en la tabla.

- Verbose Messages. Nos da el feedback de lo que esta sucediendo.

- Reindex: aplica para tablas que tienen indices, entre ellos las llaves primarias, que los motores crean como indices porque son normalmente usados para búsquedas.

- Cluster: es decirle al motor de la base de datos que reorganice el almacenamiento.

Estas 4 opciones no es recomendable hacerlo a mano ya que postgres tiene mas de 20 años optimizando estos procesos, aunque eventualmente si puede llegar a ser necesario hacerlo hazlo en un horario que no afecte a tus usuarios o servicios.

### <a name="clase-28-introduccion-a-replicas"><a/>Clase 28 Introduccion a Replicas  [Index](#index_01)

Las replicas son mecanismos que nos permiten evitar problemas  de entrada y salida en los sistemas operativos.

Ten en cuenta esta frase.

![replicas_1](src/replicas_1.jpg)

Cuando tu aplicación crece en usuarios/operaciones te vas a topar con limites en términos de física y electronica, es decir los recursos de tu servidor (disco, ram, cpu) tienen limites, ellos no pueden ser sobrepasados en sus procesos de lectura y escritura que no pueden llevarse al mismo tiempo, por este motivo postgres bloquea la tabla para escritura mientras la lees y viceversa, esto para conservar la consistencia de datos, aquí entran las replicas a solucionar este problema.

**Estrategia de Replicas**.

Tendremos la base de datos principal se hacen todas las modificaciones y tener una base de datos secundaria donde solamente se hacen las lecturas, separa las acciones nos permite separar todas las tareas que hace internamente postgres es decir los cambios los hace en "A" y  automáticamente se llevan a la tabla donde se leen los datos "B", este proceso lo hace postgres no tiene que hacer nada a mano, solamente es necesario configurar al menos 2 servidores de postgres (por lo menos 2, ya que puedes tener una cantidad ilimitada de replicas) uno como maestro y otro como esclavo, deberás modificar tu aplicación para que todas las modificaciones las haga sobre la base de datos principal o "master" y que todas las lecturas las haga sobre la replica (copia en caliente de la base de datos).

IPS (In operation per second) es la limitante a nivel de sistemas operativos, como saber que nivel nos sirve (10,20, 40 iops)

### <a name="clase-30-otras-buenas-practicas"><a/> Clase 30 Otras buenas  Practicas  [Index](#index_01)

**Problema:** Evitar bloqueos por inserciones y borrados en la misma tabla.

**solución:** renombrar tablas, decirle a sistema que ataque dos tablas, ejemplo con un **ALTER TABLE** cambia la tabla **viajes** a **viajes_temporal** y crea una tabla de **viajes "nueva"** mientras que la aplicación sigue utilizando la tabla de viajes para insertar las operaciones del dia/turno, esto permitirá a tu algoritmo de consolidación realizar el borrado y indexado mucho mas rápido, mientras la tabla viajes sigue registrando las operaciones de la aplicación, esta operación se realiza dependiendo el flujo de informacion a consolidar.
